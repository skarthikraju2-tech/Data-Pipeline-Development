```python
# Data Pipeline Development: ETL Process

This notebook automates the ETL (Extract, Transform, Load) process for data preprocessing, transformation, and loading using Pandas and Scikit-learn.
```

# Import Required Libraries

Import pandas, numpy, scikit-learn, and any other necessary libraries for ETL.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
import os
```

# Load Raw Data

Generate sample raw data using scikit-learn's make_classification for demonstration. In a real scenario, read from a CSV or database.

```python
# Generate sample data
X, y = make_classification(n_samples=1000, n_features=10, n_informative=8, n_redundant=2, n_classes=2, random_state=42)

# Create DataFrame
df = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(10)])
df['target'] = y

print("Raw data shape:", df.shape)
print(df.head())
```

# Data Preprocessing

Handle missing values, remove duplicates, and perform basic data cleaning using pandas.

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

# Basic cleaning: ensure numeric types
df = df.apply(pd.to_numeric, errors='coerce')

print("Data shape after preprocessing:", df.shape)
```

# Data Transformation

Apply transformations such as scaling and normalization using scikit-learn transformers.

```python
# Separate features and target
X = df.drop('target', axis=1)
y = df['target']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Scaled features shape:", X_scaled.shape)
```

# Feature Engineering

Create new features or modify existing ones to improve data quality and model performance.

```python
# Add polynomial features
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_scaled)

print("Polynomial features shape:", X_poly.shape)
```

# Data Splitting

Split the processed data into training and testing sets using scikit-learn's train_test_split.

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

print("Train set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

# Load Processed Data

Save the processed and transformed data to disk for downstream tasks.

```python
# Create data directory if not exists
os.makedirs('data', exist_ok=True)

# Save processed data
pd.DataFrame(X_train).to_csv('data/train_features.csv', index=False)
pd.DataFrame(y_train).to_csv('data/train_target.csv', index=False)
pd.DataFrame(X_test).to_csv('data/test_features.csv', index=False)
pd.DataFrame(y_test).to_csv('data/test_target.csv', index=False)

print("Processed data saved to data/ directory.")
```
